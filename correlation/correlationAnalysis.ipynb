```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
seaborn.set()
```

```python
# reading in the data
filtered_pos_elon_tesla_stock_data = pd.read_csv(
   "../etl_data/filtered_pos_elon_tesla_stock_data.csv"

)
filtered_neg_elon_tesla_stock_data = pd.read_csv(
    "../etl_data/filtered_neg_elon_tesla_stock_data.csv"
)
filtered_neu_elon_tesla_stock_data = pd.read_csv(
    "../etl_data/filtered_neu_elon_tesla_stock_data.csv"
)
pos_elon_tesla_stock_data = pd.read_csv("../etl_data/pos_elon_tesla_stock_data.csv")
neg_elon_tesla_stock_data = pd.read_csv("../etl_data/neg_elon_tesla_stock_data.csv")
neu_elon_tesla_stock_data = pd.read_csv("../etl_data/neu_elon_tesla_stock_data.csv")

elon_sp500_pos = pd.read_csv("../etl_data/elon_sp500_pos.csv")
elon_sp500_neg = pd.read_csv("../etl_data/elon_sp500_neg.csv")
elon_sp500_neu = pd.read_csv("../etl_data/elon_sp500_neu.csv")
```

```python
# FILTERED SP500 CHANGE PERCENT
fig, axs = plt.subplots(3, figsize=(8, 10))
fig.suptitle("Elon Tweet Sentiment Score and SP500 Change Percent Correlation")

axs[0].scatter(elon_sp500_neg["sentiment score"], elon_sp500_neg["change_percent2"])
axs[0].set_title("Negative Sentiments")

axs[1].scatter(elon_sp500_neu["sentiment score"], elon_sp500_neu["change_percent2"])
axs[1].set_title("Neutral Sentiments")

axs[2].scatter(elon_sp500_pos["sentiment score"], elon_sp500_pos["change_percent2"])
axs[2].set_title("Positive Sentiments")

for ax in axs.flat:
    ax.set(xlabel='Sentiment Score', ylabel="Change Percent")

plt.subplots_adjust(left=0.1,bottom=0.1,right=0.9, top=0.9, wspace=0.4,hspace=.4)
plt.savefig("../figures/tweetAndSP500ChangePercentCorrelation.jpg")

print(
    "pos sp500 change percent correlation: ",
    elon_sp500_pos["sentiment score"].corr(elon_sp500_pos["change_percent2"]),
)
print(
    "neg sp500 change percent correlation: ",
    elon_sp500_neg["sentiment score"].corr(elon_sp500_neg["change_percent2"]),
)
print(
    "neu sp500 change percent correlation: ",
    elon_sp500_neu["sentiment score"].corr(elon_sp500_neu["change_percent2"]),
)
```

```python
# FILTERED SP500 VOLUME
fig, axs = plt.subplots(3, figsize=(8, 10))
fig.suptitle("Elon Tweet Sentiment Score and SP500 Volume Correlation")

axs[0].scatter(elon_sp500_neg["sentiment score"], elon_sp500_neg["volume2"])
axs[0].set_title("Negative Sentiments")

axs[1].scatter(elon_sp500_neu["sentiment score"], elon_sp500_neu["volume2"])
axs[1].set_title("Neutral Sentiments")

axs[2].scatter(elon_sp500_pos["sentiment score"], elon_sp500_pos["volume2"])
axs[2].set_title("Positive Sentiments")

for ax in axs.flat:
    ax.set(xlabel='Sentiment Score', ylabel="Volume")

plt.subplots_adjust(left=0.1,bottom=0.1,right=0.9, top=0.9, wspace=0.4,hspace=.4)
plt.savefig("../figures/tweetAndSP500VolumeCorrelation.jpg")

print(
    "neg sp500 volume correlation: ",
    elon_sp500_neg["sentiment score"].corr(elon_sp500_neg["volume2"]),
)
print(
    "neu sp500 volume correlation: ",
    elon_sp500_neu["sentiment score"].corr(elon_sp500_neu["volume2"]),
)
print(
    "pos sp500 volume correlation: ",
    elon_sp500_pos["sentiment score"].corr(elon_sp500_pos["volume2"]),
)
```

```python
# UNFILTERED TESLA CHANGE PERCENT
fig, axs = plt.subplots(3, figsize=(8, 10))
fig.suptitle("Unfiltered Elon Tweet Sentiment Score and Tesla Change Percent Correlation")

axs[0].scatter(neg_elon_tesla_stock_data["sentiment score"], neg_elon_tesla_stock_data["change_percent"])
axs[0].set_title("Negative Sentiments")

axs[1].scatter(neu_elon_tesla_stock_data["sentiment score"], neu_elon_tesla_stock_data["change_percent"])
axs[1].set_title("Neutral Sentiments")

axs[2].scatter(pos_elon_tesla_stock_data["sentiment score"], pos_elon_tesla_stock_data["change_percent"])
axs[2].set_title("Positive Sentiments")

for ax in axs.flat:
    ax.set(xlabel='Sentiment Score', ylabel="Change Percent")

plt.subplots_adjust(left=0.1,bottom=0.1,right=0.9, top=0.9, wspace=0.4,hspace=.4)
plt.savefig("../figures/unfilteredTweetAndTeslaChangePercentCorrelation.jpg")

print(
    "Negative Elon Tweet Sentiment Score + Tesla Change Percent Corr.:",
    neg_elon_tesla_stock_data["sentiment score"].corr(
        neg_elon_tesla_stock_data["change_percent"]
    ),
)
print(
    "Neutral Elon Tweet Sentiment Score + Tesla Change Percent Corr.",
    neu_elon_tesla_stock_data["sentiment score"].corr(
        neu_elon_tesla_stock_data["change_percent"]
    ),
)
print(
    "Positive Elon Tweet Sentiment Score + Tesla Change Percent Corr.",
    pos_elon_tesla_stock_data["sentiment score"].corr(
        pos_elon_tesla_stock_data["change_percent"]
    ),
)
```

```python
# UNFILTERED TESLA VOLUME
fig, axs = plt.subplots(3, figsize=(8, 10))
fig.suptitle("Unfiltered Elon Tweet Sentiment Score and Tesla Volume Correlation")

axs[0].scatter(neg_elon_tesla_stock_data["sentiment score"], neg_elon_tesla_stock_data["volume"])
axs[0].set_title("Negative Sentiments")

axs[1].scatter(neu_elon_tesla_stock_data["sentiment score"], neu_elon_tesla_stock_data["volume"])
axs[1].set_title("Neutral Sentiments")

axs[2].scatter(pos_elon_tesla_stock_data["sentiment score"], pos_elon_tesla_stock_data["volume"])
axs[2].set_title("Positive Sentiments")

for ax in axs.flat:
    ax.set(xlabel='Sentiment Score', ylabel="Volume")

plt.subplots_adjust(left=0.1,bottom=0.1,right=0.9, top=0.9, wspace=0.4,hspace=.4)
plt.savefig("../figures/unfilteredTweetAndTeslaVolumeCorrelation.jpg")

print(
    "Negative Elon Tweet Sentiment Score + Tesla Volume Corr.",
    neg_elon_tesla_stock_data["sentiment score"].corr(
        neg_elon_tesla_stock_data["volume"]
    ),
)
print(
    "Neutral Elon Tweet Sentiment Score + Tesla Volume Corr.",
    neu_elon_tesla_stock_data["sentiment score"].corr(
        neu_elon_tesla_stock_data["volume"]
    ),
)
print(
    "Positive Elon Tweet Sentiment Score + Tesla Volume Corr.",
    pos_elon_tesla_stock_data["sentiment score"].corr(
        pos_elon_tesla_stock_data["volume"]
    ),
)
```

```python
# FILTERED TESLA CHANGE PERCENT
fig, axs = plt.subplots(3, figsize=(8, 10))
fig.suptitle("Filtered Elon Tweet Sentiment Score and Tesla Change Percent Correlation")

axs[0].scatter(filtered_neg_elon_tesla_stock_data["sentiment score"], filtered_neg_elon_tesla_stock_data["change_percent"])
axs[0].set_title("Negative Sentiments")

axs[1].scatter(filtered_neu_elon_tesla_stock_data["sentiment score"], filtered_neu_elon_tesla_stock_data["change_percent"])
axs[1].set_title("Neutral Sentiments")

axs[2].scatter(filtered_pos_elon_tesla_stock_data["sentiment score"], filtered_pos_elon_tesla_stock_data["change_percent"])
axs[2].set_title("Positive Sentiments")

for ax in axs.flat:
    ax.set(xlabel='Sentiment Score', ylabel="Change Percent")

plt.subplots_adjust(left=0.1,bottom=0.1,right=0.9, top=0.9, wspace=0.4,hspace=.4)
plt.savefig("../figures/filteredTweetAndTeslaChangePercentCorrelation.jpg")

print(
    "positive filtered change percent: ",
    filtered_pos_elon_tesla_stock_data["sentiment score"].corr(
        filtered_pos_elon_tesla_stock_data["change_percent"]
    ),
)
print(
    "Negative Filtered tweet change percent + sentiment score corr",
    filtered_neg_elon_tesla_stock_data["sentiment score"].corr(
        filtered_neg_elon_tesla_stock_data["change_percent"]
    ),
)
print(
    "Neutral Filtered tweet change percent + sentiment score corr",
    filtered_neu_elon_tesla_stock_data["sentiment score"].corr(
        filtered_neu_elon_tesla_stock_data["change_percent"]
    ),
)
```

```python
# FILTERED TESLA VOLUME
fig, axs = plt.subplots(3, figsize=(8, 10))
fig.suptitle("Filtered Elon Tweet Sentiment Score and Tesla Volume Correlation")

axs[0].scatter(filtered_neg_elon_tesla_stock_data["sentiment score"], filtered_neg_elon_tesla_stock_data["volume"])
axs[0].set_title("Negative Sentiments")

axs[1].scatter(filtered_neu_elon_tesla_stock_data["sentiment score"], filtered_neu_elon_tesla_stock_data["volume"])
axs[1].set_title("Neutral Sentiments")

axs[2].scatter(filtered_pos_elon_tesla_stock_data["sentiment score"], filtered_pos_elon_tesla_stock_data["volume"])
axs[2].set_title("Positive Sentiments")

for ax in axs.flat:
    ax.set(xlabel='Sentiment Score', ylabel="Volume")

plt.subplots_adjust(left=0.1,bottom=0.1,right=0.9, top=0.9, wspace=0.4,hspace=.4)
plt.savefig("../figures/filteredTweetAndTeslaVolumeCorrelation.jpg")

print(
    "positive filtered volume: ",
    filtered_pos_elon_tesla_stock_data["sentiment score"].corr(
        filtered_pos_elon_tesla_stock_data["volume"]
    ),
)
print(
    "Negative Filtered elon tweet volume + sentiment score corr",
    filtered_neg_elon_tesla_stock_data["sentiment score"].corr(
        filtered_neg_elon_tesla_stock_data["volume"]
    ),
)
print(
    "Neutral Filtered elon tweet volume + sentiment score corr",
    filtered_neu_elon_tesla_stock_data["sentiment score"].corr(
        filtered_neu_elon_tesla_stock_data["volume"]
    ),
)
```
